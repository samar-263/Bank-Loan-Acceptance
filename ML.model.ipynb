```python
import numpy as np
import pandas as pd
import warnings
warnings.filterwarnings("ignore")
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv(r'C:\Users\samar\OneDrive\Desktop\DEFAULT\DEFAULTER RECOGNIZER FOR BANK LOAN.csv')
data
```

```python
dep_col = ['TARGET', 'AMT_INCOME_TOTAL','DAYS_BIRTH', 'DAYS_EMPLOYED']
```

```python
dep_data = data[dep_col]
dep_data
```

```python
dep_data = pd.get_dummies(dep_data)
dep_data
```

```python
X  = dep_data.iloc[:,1:]
y = dep_data.iloc[:,:1]
```

```python
X.head() 
```

```python
from sklearn.model_selection import train_test_split
X_train ,X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

Here i just checked  according to my dataset which model is suitable for buliding ml model.

```python
from sklearn.ensemble import RandomForestClassifier
cls = RandomForestClassifier(n_estimators=2000,criterion='entropy')
cls.fit(X_train,y_train)
```

```python
y_pred = cls.predict(X_test)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
from sklearn.metrics import accuracy_score
ac = accuracy_score(y_test, y_pred)
ac
```
